```python
import os, random, numpy as np, pandas as pd
import tensorflow as tf
from transformers import AutoTokenizer, TFAutoModel
from sklearn.utils import resample
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras import layers, models
from tensorflow.keras.layers import TextVectorization

def seed_everything(seed=42):
    random.seed(seed); np.random.seed(seed); tf.random.set_seed(seed)
seed_everything(42)

save_dir = "./marketmind_finbert_artifacts_full"
os.makedirs(save_dir, exist_ok=True)

csv_path = "marketmind_dataset_part1.csv"
df = pd.read_csv(csv_path)
print("[INFO] Columns:", list(df.columns))
print(df.head(3))

text_col, label_col = "NewsOrComment", "Recommendation"
extra_cols = ["ReasonForIncreaseOrDecrease", "Status"]

df["_text"] = df[text_col].fillna("").astype(str)
for c in extra_cols:
    if c in df.columns:
        df["_text"] += " " + df[c].fillna("").astype(str)

df[label_col] = df[label_col].astype(str).str.strip().str.lower()
label_map = {"sell": 0, "hold": 1, "buy": 2}
df = df[df[label_col].isin(label_map.keys())]
df["_yid"] = df[label_col].map(label_map)
print(f"[INFO] Label distribution:\n{df['_yid'].value_counts()}")

max_n = df["_yid"].value_counts().max()
df_bal = pd.concat([
    resample(df[df["_yid"] == k], replace=True, n_samples=max_n, random_state=42)
    for k in df["_yid"].unique()
])
df = df_bal.sample(frac=1.0, random_state=42).reset_index(drop=True)
print(f"[INFO] After balancing:\n{df['_yid'].value_counts()}")

n = len(df)
train_size, val_size = int(0.7*n), int(0.15*n)
train_df = df.iloc[:train_size]
val_df   = df.iloc[train_size:train_size+val_size]
test_df  = df.iloc[train_size+val_size:]
print(f"[INFO] Split sizes → Train={len(train_df)}, Val={len(val_df)}, Test={len(test_df)}")

print("[INFO] Loading FinBERT model...")
tokenizer = AutoTokenizer.from_pretrained("yiyanghkust/finbert-tone")
bert = TFAutoModel.from_pretrained("yiyanghkust/finbert-tone")

def encode_texts(texts, tokenizer, max_len=128):
    return tokenizer(
        texts.tolist(),
        truncation=True,
        padding="max_length",
        max_length=max_len,
        return_tensors="tf"
    )

train_enc = encode_texts(train_df["_text"], tokenizer)
val_enc   = encode_texts(val_df["_text"], tokenizer)
test_enc  = encode_texts(test_df["_text"], tokenizer)

y_train, y_val, y_test = train_df["_yid"].values, val_df["_yid"].values, test_df["_yid"].values

input_ids = tf.keras.Input(shape=(None,), dtype=tf.int32, name="input_ids")
mask      = tf.keras.Input(shape=(None,), dtype=tf.int32, name="attention_mask")

def bert_encoder(inputs):
    ids, attn = inputs
    outputs = bert(ids, attention_mask=attn, training=False)
    return outputs.last_hidden_state[:, 0, :]  # CLS token

bert_out = tf.keras.layers.Lambda(bert_encoder, name="bert_encoder", output_shape=(768,))([input_ids, mask])
x = tf.keras.layers.Dense(256, activation="relu")(bert_out)
x = tf.keras.layers.Dropout(0.3)(x)
logits = tf.keras.layers.Dense(3, activation="softmax")(x)

model = tf.keras.Model(inputs=[input_ids, mask], outputs=logits)
model.compile(optimizer=tf.keras.optimizers.Adam(3e-5),
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])
print(model.summary())

hist = model.fit(
    {"input_ids": train_enc["input_ids"], "attention_mask": train_enc["attention_mask"]},
    y_train,
    validation_data=(
        {"input_ids": val_enc["input_ids"], "attention_mask": val_enc["attention_mask"]},
        y_val),
    epochs=10,
    batch_size=8,
    verbose=2
)

plt.figure()
plt.plot(hist.history["accuracy"], label="train_acc")
plt.plot(hist.history["val_accuracy"], label="val_acc")
plt.legend(); plt.title("Training Curves (FinBERT, 10 Epochs)")
plt.xlabel("Epochs"); plt.ylabel("Accuracy")
plt.savefig(os.path.join(save_dir, "training_curves.png"), dpi=150)
plt.close()

print("\n[INFO] Evaluating on test set...")
probs = model.predict({"input_ids": test_enc["input_ids"],
                       "attention_mask": test_enc["attention_mask"]}, verbose=0)
y_pred = probs.argmax(axis=1)

acc = accuracy_score(y_test, y_pred)
f1  = f1_score(y_test, y_pred, average="macro")
print(f"\n[Test Results] Accuracy={acc:.3f}, Macro-F1={f1:.3f}")
print(classification_report(y_test, y_pred, digits=4))

cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d",
            xticklabels=["Sell","Hold","Buy"],
            yticklabels=["Sell","Hold","Buy"])
plt.title("Confusion Matrix (FinBERT Test)")
plt.xlabel("Predicted"); plt.ylabel("Actual")
plt.tight_layout()
plt.show()
plt.savefig(os.path.join(save_dir, "confusion_matrix.png"), dpi=150)
plt.close()
print("[INFO] Saved confusion_matrix.png")

plt.figure()
plt.plot(hist.history["accuracy"], label="Train Accuracy", marker='o')
plt.plot(hist.history["val_accuracy"], label="Validation Accuracy", marker='o')
plt.legend()
plt.title("Accuracy vs Epochs (FinBERT)")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.grid(True)
plt.tight_layout()
plt.show()
plt.savefig(os.path.join(save_dir, "accuracy_vs_epochs.png"), dpi=150)
plt.close()

plt.figure()
plt.plot(hist.history["loss"], label="Train Loss", marker='s')
plt.plot(hist.history["val_loss"], label="Validation Loss", marker='s')
plt.legend()
plt.title("Loss vs Epochs (FinBERT)")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.grid(True)
plt.tight_layout()
plt.show()
plt.savefig(os.path.join(save_dir, "loss_vs_epochs.png"), dpi=150)
plt.close()

print("[INFO] Accuracy and loss graphs displayed and saved!")

inv_map = {v:k for k,v in label_map.items()}
test_df["PredictedLabel"] = [inv_map[i] for i in y_pred]

print("\n[Sample Predictions]")
preview_cols = ["CompanyName", "ArticleDate", "NewsOrComment", "Recommendation", "PredictedLabel"]
pd.set_option("display.max_colwidth", 120)
sample_display = test_df.sample(min(10, len(test_df)), random_state=42)[preview_cols]
print(sample_display.to_string(index=False))

model.save(os.path.join(save_dir, "marketmind_finbert_model.keras"))
print("\nTraining complete. Artifacts saved to:", save_dir)
```

```python
import os
import socket
import numpy as np
import gradio as gr
import tensorflow as tf
from datetime import datetime
from transformers import AutoTokenizer, TFAutoModel
from keras.saving import register_keras_serializable
from openai import OpenAI

KERAS_MODEL_PATH = "./marketmind_finbert_artifacts_full/marketmind_finbert_model.keras"
TOKENIZER_NAME = "yiyanghkust/finbert-tone"
OPENAI_MODEL = "gpt-4o-mini"
label_map = {0: "Sell", 1: "Hold", 2: "Buy"}

def get_openai_client():
    key = os.getenv("OPENAI_API_KEY")
    if not key:
        raise RuntimeError("Set OPENAI_API_KEY first!")
    return OpenAI(api_key=key)

client = get_openai_client()

def extract_company_with_gpt(text):
    resp = client.chat.completions.create(
        model=OPENAI_MODEL,
        messages=[
            {
                "role": "system",
                "content": "Extract only ONE company name from the message. Return ONLY the name."
            },
            {"role": "user", "content": text}
        ],
        temperature=0,
        max_tokens=10
    )
    return resp.choices[0].message.content.strip()


def rewrite_headline(company, user_input):
    current_month = datetime.now().strftime("%B %Y")
    system_msg = (
        f"Convert this investor query into a short factual headline. "
        f"Use the current month and year: {current_month}. "
        f"Headline should mention: {company}. "
        f"No advice, no made-up dates.\n\n"
    )
    resp = client.chat.completions.create(
        model=OPENAI_MODEL,
        messages=[
            {"role": "system", "content": system_msg},
            {"role": "user", "content": user_input},
        ],
        temperature=0.2,
        max_tokens=40,
    )
    return resp.choices[0].message.content.strip()

tokenizer = AutoTokenizer.from_pretrained(TOKENIZER_NAME)
bert_base = TFAutoModel.from_pretrained(TOKENIZER_NAME)

@register_keras_serializable(package="MarketMind")
def bert_encoder(inputs):
    ids, attn = inputs
    return bert_base(ids, attention_mask=attn, training=False).last_hidden_state[:, 0, :]

model = tf.keras.models.load_model(
    KERAS_MODEL_PATH,
    custom_objects={"bert_encoder": bert_encoder, "TFBertModel": bert_base.__class__}
)

def finbert_predict(text):
    enc = tokenizer([text], truncation=True, padding="max_length",
                    max_length=128, return_tensors="tf")
    probs = model.predict({
        "input_ids": enc["input_ids"],
        "attention_mask": enc["attention_mask"]
    }, verbose=0)[0]
    idx = int(np.argmax(probs))
    return label_map[idx], probs

def investment_chat(request, history):
    history = history or []
    company_now = extract_company_with_gpt(request)
    company_prev = extract_company_with_gpt(history[-1][0]) if history else None
    combined = request if company_now != company_prev else history[-1][1] + " " + request

    company = company_now or company_prev or "the company"
    headline = rewrite_headline(company, combined)
    sentiment, probs = finbert_predict(headline)

    reply = (
        f"📊 **Market Insight:** {headline}\n"
        f"**Action → {sentiment}**\n"
        f"- Sell: {probs[0]*100:.1f}%\n"
        f"- Hold: {probs[1]*100:.1f}%\n"
        f"- Buy:  {probs[2]*100:.1f}%"
    )

    history.append((request, reply))
    return history, history, ""

def free_port():
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as s:
        s.bind(("", 0))
        return s.getsockname()[1]

css = """
.gradio-container {
    background: #101418 !important;
    color: white !important;
}

h2 {
    color: white !important;
}

.gr-chatbot {
    background: #1a1f24 !important;
    border-radius: 12px !important;
    border: 1px solid #2c343a !important;
}

textarea, input {
    background: #1a1f24 !important;
    color: white !important;
    border: 1px solid #2c343a !important;
    border-radius: 10px !important;
}

.user {
    background: #3b82f6 !important;
    color: white !important;
    border-radius: 12px !important;
}

.bot {
    background: #2a2f33 !important;
    color: #e2e2e2 !important;
    border-radius: 12px !important;
}
"""

with gr.Blocks(css=css, title="MarketMind AI") as demo:
    gr.Markdown("<h2 style='text-align:center;'>🤖 MarketMind AI Advisor</h2>")
    gr.Markdown(
        """
        <div style='text-align:center; font-size:16px; color:#d1d5db;'>
        Ask naturally about companies or investment contexts.<br>
        <b>🚀 Examples:</b><br>
        • <i>"Hi, I want to invest in Tesla. It’s November and stocks are rising."</i><br>
        • <i>"I'm thinking about Amazon this spring season."</i><br>
        • <i>"I live in India and want to invest in Rivian."</i><br>
        </div>
        """
    )

    chat = gr.Chatbot(height=450, show_label=False)
    input_box = gr.Textbox(
        placeholder="Ask a stock question…", show_label=False
    )

    input_box.submit(
        investment_chat,
        inputs=[input_box, chat],
        outputs=[chat, chat, input_box]
    )

if __name__ == "__main__":
    port = free_port()
    print(f"Running on: http://127.0.0.1:{port}")
    demo.launch(server_name="0.0.0.0", server_port=port)
```
